# Cálculo do RMC do Sistem Solar

Este notebook realiza o cálculo do Radial Mass Concentration (RMC) do Sistema Solar de acordo com o artigo de Chambers, 2001

$$\text{RMC} = \text{max}\left(\frac{\sum{M_j}}{\sum{M_j} [\log_{10}(a / a_j)]^2}\right)$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = np.array([0.0553, 0.815, 1.0, 0.107])
a = np.array([0.387, 0.723, 1.0, 1.52])
range_ss = np.linspace(0.1, 1.6, 1000)
```

```python
sum_mass = np.sum(m)
sum_den = 0.0
for i in range(len(m)):
    sum_den = sum_den + m[i] * (np.log10(range_ss/a[i]))**2
```

```python
RMC_list = sum_mass/sum_den 
RMC = max(RMC_list)
print('Valor de RMC = {0:0.1f}'.format(RMC))
```

```python
plt.figure(figsize=(5,4))
plt.xlabel('Semimajor axis [au]')
plt.ylabel('RMC')
plt.plot(range_ss, RMC_list)
plt.scatter(a, [0]*len(a), s=m*500)
```
